1) Relacionado con la naturaleza de los datos, qué condiciones deben cumplir las matrices $A$ y $b$ para ser resolubles desde la perspectiva de los mínimos cuadrados.

**Lema**

Sea $A\in \mathbb{R}^{n\times m}$  si las columnas de $A$ forman un conjunto linealmente independiente, entonces existe una descomposición QR de $A$, es decir, $A$ se puede escribir como el producto de una matriz ortogonal  $Q$, con una matriz triangular superior  con numeros positivos en la digonal $R$. 

$desmostración$

denominemos $v_1,v_2,\dots,v_m$ a las columnas de $A$. Por hipótesis $B=\{v_1,v_2,\dots,v_m\}$ es un conjunto linealmente independiente. Aplicando el procedimineto de Gram-Schmidt al conjunto B obtenemos una base ortognal $u_1,\dots,u_m$ de $col(A)$ que satisface las igualdades

$u_1=v_1$

$u_2=v_2-\alpha_{12}u_1$

$\dots$

$u_m=v_m-\alpha_{1m}u_1-\alpha_{2m}u_2-\dots-\alpha_{(m-1)m}u_{m-1}$

aqui $\alpha_{ij}=\cfrac{u_i^{H}v_j}{||u_i||^2}$

Pero entonces, despejando cada $v_i$ obtenemos las igualdades

$v_1=u_1$

$v_2=\alpha_{12}u_1+u_2$

$\dots$

$v_m=\alpha_{1m}u_1+\alpha_{2m}u_2+\dots+\alpha_{(m-1)m}u_{m-1}+u_m$

que puede escribirse en forma matricial como 

$$[v_1 v_2 \cdots v_m]=[u_1 u_2 \cdots u_m]\left[\begin{smallmatrix}
1 & \alpha_{12} & \alpha_{13} & \dots & \alpha_{1m}\\
0 & 1 & \alpha_{23} & \dots & \alpha_{2m}\\
0 & 0 & 1 & \dots & \alpha_{3m}\\
\vdots & \vdots & \vdots & \ddots & \vdots\\
0 & 0 & 0 & \dots & 1
\end{smallmatrix}\right]$$

 Llamando $Q_0=[u_1,\dots, u_m]$ y $R_0$ a la matriz triangular superior que aparece arriba, tenemops que $A=Q_0R_0$ que es casi la factorización que estamos buscando, ya que las columnas de $Q_0$ forman un conjunto ortogonal y $R_0$ es triangular superior e inversible. Lo que hacemos ahora es normalizar cada columna de $Q_0$, es decir, definamos 
 
 $$Q=[q_1 q_2 \cdots q_m]$$ con $$ q_i=\cfrac{u_i}{||u_i||}$$
 
 y modificamos $R_0$ de modo tal que su producto con $Q$ siga dando $A$, para ello multiplicamos cada fila de $R_0$ por el número por el cual dividimo la correspondiente coluna de $Q_0$ obteniendo que 
 
 $$R=\left[\begin{smallmatrix}
||u_1|| & \alpha_{12}||u_1|| & \alpha_{13}||u_1|| & \dots & \alpha_{1m}||u_1||\\
0 & ||u_2|| & \alpha_{23}||u_2|| & \dots & \alpha_{2m}||u_2||\\
0 & 0 & ||u_3|| & \dots & \alpha_{3m}||u_3||\\
\vdots & \vdots & \vdots & \ddots & \vdots\\
0 & 0 & 0 & \dots & ||u_m||
\end{smallmatrix}\right]$$

Así $A=QR$, la factorización buscada. 

Note que con lo anterior $A^TA=(QR)^T(QR)=R^TQ^TQR=R^TR$, como $R$ es invertible por ser una matriz superior con toda su diagonal principal no nula, entonces $A^TA=R^TR$ es invertible.




**Teorema**

Sea $A$ una matriz $n\times m$, y suponga que tenemos el sistema $Ax=b$, entonces el problema de mínimos cuadrados (minimizar $||Ax-b||_2$) tiene solución única y viene dada por $x=A^{+}b$ (donde $A^{+}=(A^tA)^{-1}A^t$), si las columnas de $A$ son linealmente independientes y $b$ es un vecotor de tamaño $n\times 1$.


$desmostración$

Dado que $A$ tiene sus columnas linealmente independientes, entonces se tiene que $(A^tA)^{-1}$ existe, ahora el problema de minimos cuadrados se puede ver como la minimizacion de la función $$J(x) = \| Ax - b\|_2^2 = ( Ax - b )^T ( Ax - b ) = (A x)^T (A x) - (A x)^T b - b^T (A x) + b^T b= x^T A^T Ax - 2 x^T A^T b + b^T b$$, luego para encontrar el minimo de esta función basta encontrar cuando la derivada parcial de $J$, respcto a $x$ es 0, esto es cuando 
$$\frac{\partial J(x)}{\partial  x} = - 2 A^T b + 2 A^T A x = 0$$
a partir de esto obtenemos la ecuación 
$A^T A x = A^T b$, ahora bien utilizando el hecho de que $(A^TA)^{-1}$ existe (por $A$ tener su columnas l.i.) , obtenemos sacando inversa a ambos lados que $x=(A^TA)^{-1}A^T b$. Dado que la psudoinversa de una matriz $n\times m$ es única (https://en.wikipedia.org/wiki/Moore%E2%80%93Penrose_inverse#Definition), y $A^{+}=(A^TA)^{-1}A^T$ cumple los 4 requerimentos para ser una psudoinversa $(AA^{+}A, A^+A^+=A^+, (AA^+)^T=AA^+, (A^+A)^T=A^+A)$, se sigue que la solución a este problema es unica y biene dada por  $x=A^{+}b$.


De esta forma las unicas condiciones sobre los datos que se necesitan para que  el problema de minimos cuadrados sea resoluble, es que las columnas de $A$ sean linealmente independites y que que $b$ tenga tamaño adecuado para que sea pueda efectuar la operación $Ax-b$. 
Como nota se puede demostrar que si $A$ no tiene columnas linealmente independientes, entonces pueden haber una infinidad de solucuciones dadas por $x=A^{+}b+(I-A^{+}A)w$ para cualquier vector $w$ (https://en.wikipedia.org/wiki/Moore%E2%80%93Penrose_inverse#Definition)



2) Use mínimos cuadrados para modelos no lineales, ajuste el precio de bitcoin y realice una predicción usando el modelo para determinar el precio diario futuro con una semana de anticipación, compárelo con los valores reales.

(para la solución de este problema utilizamos minimos cuadrados lineales, para ajustar un problema no-lineal, y no mininimos cuadrados nolineales) mprimero imprimamos los datos que necesitaremos para plantear nuestro problema para hallar la ecuación lineal que mejor se aproxime a estos.

```julia
using CSV
using DataFrames

dataBitcoin = CSV.read("dataBitcoin.csv", DataFrame)

#En la primera columna esta impresa la fecha, en la segunda columna esta impresa el último valor que se registro del bitcoin en
# la fecha dada, en la tercera columna esta impreso el valor del bitcoin al iniciar el dia, y en las cuarta y quinta columna estan
#impresos los valores máximo y mínimo del bitcoin ese dia (todo esto en dolares).  


```

Acontinuación veremos el valor del bitcoin en los últimos 29 días. comenzando desde el 19/02/2022.(datos tomados de: https://es.investing.com/crypto/bitcoin/historical-data)

```julia
using Plots
using CSV
using DataFrames

dias = collect(29:-1:1)

plot(dias,dataBitcoin.Último, title="Valor del Bitcoin en los ultimos 29 días",label= "valor",xlabel="días", ylabel="valor")
```

 Ahora formulamos quien es $A$, $b$ en el problema de minimos cuadrados, en este caso planteamos las ecuaciones $mx_i+d=y_i$, donde $x_i$, es el $i$-ésimo día, y $y_i$ es el precio en el $i$-ésimo día, esto plantea un sistema de ecuaciones, donde $A$, es la matriz que en su primera columna tiene los dias, y en la segunda columna tiene unos, donde los parametros a encontrar son la pendiente y el corte con el eje x, y el vector $b$ son los $y_i$.

```julia
N = size(dataBitcoin.Último,1); #tamaño de la matriz, pero solo las filas (si se coloca 2 nos da las columnas)
A = Array(1:N);                 #se crea un vector columna que de 1 hasta N                   
b = Array(dataBitcoin.Último);
A = [dias ones(N,1)];
```

 Utilizando descomposición $QR$,(Donde $Q$ es una matriz ortogonal y $R$ es una matriz diagonal superor), podemos resolver la ecuación $x=(A^TA)^{-1}A^T$ que como vimos anteriormente nos da la solución a nuestro problema, sabemos además que $A$ se puede descomponer de esta forma si $A$ tiene sus columnas linealmente independientes, lo cual sucede en este caso (puesto que no existen escalares distintos de 0 $c_1,c_2$ tales que $c_1 i+c_2=0$ y $c_1 i+c_2=0$ para $i=1,2,\dots 29$.). cambiando $A$ por $QR$, en la ecuaicon $x=(A^TA)^{-1}A^T$, obtenemos $x=((QR)^TQR)^{-1}(QR)^T=(R^TQ^TQR)R^TQ^Tb=(R^TR)^{-1}R^TQ^Tb=R^{-1}R^{-T}RQ^Tb=R^{-1}Q^Tb$, así el sistema que tenemos que solucionar es $Rx=Q^Tb$

```julia
using LinearAlgebra

qrA = qr(A);                    # descomposicion de A en su forma QR
solution= qrA\b                 # "\" resuelve el sitema QRx=b.
```

Obtenemos así la solución $m=60.130049261082696$ y $d=38908.42857142858$, graficandola obrentendiramos lo siguiente 


```julia
using Plots


g(x)=solution[1]x+solution[2];

plot(dias,dataBitcoin.Último, title="Valor del Bitcoin en los ultimos 29 días",label= "valor",xlabel="días", ylabel="valor");

plot!(g, title="Valor del Bitcoin en los ultimos 29 días",label="aproximación",xlabel="días", ylabel="valor")



```

```julia
#Según nuestro modelo tenemos las siguientes predicciónes para los siguinetes días.

using Printf
@printf("20 de marzo=%0.1f ", float(g(30)))
@printf(" 21 de marzo=%0.1f", float(g(31)))
@printf(" 22 de marzo=%0.1f", float(g(32)))
@printf(" 23 de marzo=%0.1f", float(g(33)))
@printf(" 24 de marzo=%0.1f", float(g(34)))
@printf(" 25 de marzo=%0.1f", float(g(35)))
@printf(" 26 de marzo=%0.1f", float(g(36)))
```

3) Use mínimos cuadrados para determinar la relación entre el precio de bitcoin y los precios de tres materias primas, a saber, petróleo, plata y oro, durante el último año.

Los datos de los precios del bitcoin, petroleo, oro y plata (en dolares) en el íltimo año se muestran acontinuación.(datos tomados de (https://es.investing.com/)

```julia
using CSV
using DataFrames
using Plots

todo = CSV.read("todo.csv", DataFrame)
p1=plot(todo.Bitcoin,label= "Bitcoin",xlabel="días", ylabel="valor");
p2=plot(todo.Petroleo,label="petroleo",xlabel="días", ylabel="valor");
p3=plot(todo.oro,label="oro",xlabel="días", ylabel="valor");
p4=plot(todo.plata,label="plata",xlabel="días", ylabel="valor");

plot(p1,p2,p3,p4)
```

Ahora formulamos quien es $A$ y $b$, en el problema de minimos cuadrados, planteemos primereo las siguientes ecuaciones lineales $a_1x_i+a_2y_i+a_3w_i=z_i$, donde $x_i$ es el valor del petroleo en el $i$-ésimo día, $y_i$ es el valor del oro en el $i$-ésimo día, $w_i$ es el valor de la plata en el $i$-ésimo día y $z_i$ es el valor del bitcoin en el $i$-ésimo día, donde $i=1,\dots, 366$ y todo esto en dolares, una vez planteadas esas ecuaciones la podemos ver como $Ax=b$, donde $A$ es la matriz que en su primera columna tiene los valores del petroleo, en su segunda columna tiene los valores del oro, y en su tercera columna tiene los valores de la plata. $x$ es el vector de parametros a encontrar y $b$ es el vector que tiene los valores del bitcoin.

```julia
N = size(todo.plata,1); #tamaño de la matriz, pero solo las filas (si se coloca 2 nos da las columnas)
                 #se crea un vector columna que de 1 hasta N                   
b = todo.Bitcoin;
A = [todo.Petroleo todo.oro todo.plata];


```

```julia
#utilizando factorizacion QR, para solucionar nuestro problema obtenemos.

using LinearAlgebra



qrA = qr(A);                    # descomposicion de A en su forma QR
solution= qrA\b
```

De esta forma, ya tenemos una función lienal que aproxima el precio del bitocioin dados los datos del petroleo, plata y oro, la cual es: $f(x,y,w)= -68.08007734709437x+23.182485858955783y+402.70981633446456w$
